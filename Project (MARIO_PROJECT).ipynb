```python
class character:
    def __init__(self,name):
        self.name=name
        self.life=3
        self.score=0

    def kicked(self):
        self.score = self.score+10
        print(f"kicked point(10) :{self.score}")

    def punched(self):
        self.score=self.score+5
        print(f"punched point(5) :{self.score}")

    def stabbed(self):
        self.life=self.life-1
        print(f"left life : {self.life}")
        if(self.life==0):
            print("game over")

    def displaylife(self):
        print(f"current life :{self.life}")
        

    def displayscore(self):
        print(f"total score : {self.score}")

        
n =input("enter a player name :")
mario = character(n)

while True:
    print(f""" Ready to play {n}
          1.kicked
          2.Punched
          3.Stabbed
          4.Display life
          5.Display score
          6.Quit game""")
    x=input("Enter a action :")
    if x=="1":
        mario.kicked()
       
    elif x=="2":
        mario.punched()

    elif x=="3":
        mario.stabbed()
        
    elif x=="4":
        mario.displaylife()

    elif x=="5":
        mario.displayscore()

    elif x=="6":
        print("Well Try")
        break
```
